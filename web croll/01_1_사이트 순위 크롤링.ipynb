### 이용 빈도가 높은 웹사이트 순위를 제공하는 사이트 크롤링

- 크롤링 사이트
    -https://www.alexa.com/topsites/countries/KR


```python
import requests
import bs4
from bs4 import BeautifulSoup
```

```python
url="https://www.alexa.com/topsites/countries/KR"
    
html_website_ranking=requests.get(url).text
# html_website_ranking[:2000] 잘 넘어오는지 임의로 확인하기 위해서.
# 웬만한 사이트는 get 방식으로 처리가능.post나 보안형태는 적은편
```

```python
soup_website_ranking=BeautifulSoup(html_website_ranking,"html.parser") #파서기를 lxml로 사용해도됨.오히려 이게 더 활용이 많은편

soup_website_ranking
```

```python
# p태그의 요소 안에서 a태그의 요소를 찾음

website_ranking=soup_website_ranking.select("p a")
                                           
website_ranking[0:6] # 첫번째 a태그에 순위와 상관없는 다른 자료가 들어있음
```

```python
print(type(website_ranking[0]))
print(website_ranking[0].txt) # 태그 내부의 text를 저장하고 있는 속성
print(website_ranking[0].get_txt) # 태그 내부의 text를 반환하는 함수
```

```python
# for문으로 하려면
lis=[]
for elem in website_ranking:
    lis.append(elem.get_txt())
lis
```

```python
# 모든 랭킹을 가져와서 리스트에 담기
website_ranking_domain= [website_ranking_element.get_text() for website_ranking_element in website_ranking]
print(website_ranking_domain[0:6])
print(len(website_ranking_domain))
```

```python
# 순위와 상관없는 첫번째 요소 제거
# website_ranking_domain.pop(0)  #'this explanation' 제거
print(website_ranking_domain[0:6])
```

```python
print(len(website_ranking_domain))
```

```python
# 전체 data 출력 후 확인
print("[Top site in south korea]")
for k in range(len(website_ranking_domain)):
    print("{0} : {1}".format(k+1,website_ranking_domain[k]))
```
