#### 동적 로그인 진행
- 프로그램코드를 통해서 자동 로그인되게 브라우저를 제어

````python
# webdriver 객체 생성
driver =webdriver.Chrome(".\driver\chromedriver")
# 페이지 접근
url ='https://nid.naver.com/nidlogin.login' 
driver.get(url)
````

````python
# id 값을 전달
# 1. id를 입력할 element(input tag)를 찾아야 함- input type="text"  id="id"
id_elem=driver.find_element_by_id("id")

# 2. 미리 입력되어있는 값이 잇을수도 있으므로 지우기
id_elem.clear()   # 아이디값 지우기

# 3. send_keys() 함수를 이용해서 element 에 값을 전달
id_elem.send_keys("taropoint1")  #입력할 아이디 입력
#참고로 session이 종종 끊기는 경우가 생긴다.그러면 걍 웹드라이버접근부터 다시하삼
````

````python
# password 값을 전달

# # 1. password를 입력할 element(input tag)를 찾아야 함- input type="text"  id="pw"
pw_elem=driver.find_element_by_id("pw")

# 2. 미리 입력되어있는 값이 잇을수도 있으므로 지우기
pw_elem.clear()   # 아이디값 지우기


# 3. send_keys() 함수를 이용해서 element 에 값을 전달
pw_elem.send_keys("sosabul0313!!")  #입력할 아이디 입력
#참고로 session이 종종 끊기는 경우가 생긴다.그러면 걍 웹드라이버접근부터 다시하삼
````

````python
# 로그인 버튼 클릭
# 개발자 코드에서 copy후 xpath 찾기함
xpath='//*[@id="log.login"]'

driver.find_element_by_xpath(xpath).click()


# 여기까지 완료되면 자동입력 방지 문자를 입력하는 기능이 나타남 
````
